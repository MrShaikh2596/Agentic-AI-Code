### Basic Agent

```python
from crewai import Agent, Crew,Task,LLM
from crewai_tools import SerperDevTool
```

```python
#setting the API keys
from google.colab import userdata
import os
os.environ["SERPER_API_KEY"] = userdata.get('SERPER_API_KEY')
os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')
os.environ["HUGGING_FACE_API_TOKEN"] = userdata.get('HUGGING_FACE_API_TOKEN')
```

```python
#definging the LLM
groq_llm= LLM(model="groq/llama-3.1-70b-versatile")
```

```python
#defining the tools
web_search = SerperDevTool()
```

```python
#Creating first Agent
Farming_Researcher = Agent(
    role="A Farming Researcher",
    goal= "To provid fine-grained and information on a {topic} given by user",
    backstory="you are a farmer with 40+ years of experience in farming various crops"
              "Most of these crops you have grown in Maharashtra, India"
              "Using you experience you provide the detailed fine-grained information about the {topic} after researching on it through internet",
    llm=groq_llm,
    respect_context_window = True
)
```

```python
#Creating first task
Farm_Task = Task(
    description="Extract the topic from the input, construct a relevant search query, and use the web_search tool to find detailed information about the topic",
    tools=[web_search],
    expected_output= "A detailed summay after fine-grained research.Also write the output to a markdown file",
    output_file="Alphonso.md",
    agent=Farming_Researcher
)
```

```python
#creating crew
Farm_Crew = Crew(
    agents=[Farming_Researcher],
    tasks=[Farm_Task],
    verbose=True
)
```

```python
topic_search = {'topic':"Alphonso Farming"}
```

```python
 Farm_Crew.kickoff(inputs=topic_search)
```

### Text-to-Speech

```python
 #using a model to convert text to speech
text_speech = LLM(model="suno/bark",model_kwargs={"provider": "huggingface"})
```

```python
Story_Agent = Agent(
    role="Story Teller",
    goal = "You are a smart agent which uses provided LLM to convert the given text into speech",
    backstory="based on your experties of Text-to-Specch capabilities you convert the given {story} into sppech "
              "using the model provided you for doing the taks",
    model=text_speech,
    respect_context_window = True
)
```

```python
Story_Task=Task(
    description="Utilize the Agent to generate the audio file of the given {story} file",
    expected_output="generate an audion file based on the given story",
    output_file="Story.mp3",
    agent=Story_Agent
)
```

```python
Story_Crew = Crew(
    agents=[Story_Agent],
    tasks=[Story_Task],
    verbose=True
)
```
