```python
from google.colab import userdata
LANGCHAIN_API_KEY = userdata.get('LANGCHAIN_API_KEY')
```

```python
#importing LLM from Hugging Face
from langchain_huggingface import HuggingFaceEndpoint
import os
os.environ["LANGCHAIN_ENDPOINT"] = "https://api.smith.langchain.com"
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_API_KEY"] = LANGCHAIN_API_KEY
os.environ['LANGCHAIN_PROJECT'] = 'Lang-Graph'
os.environ['HUGGINGFACEHUB_API_TOKEN'] = userdata.get('HUGGINGFACEHUB_API_TOKEN')
```

```python
#importig model from HF
llm = HuggingFaceEndpoint(repo_id="Qwen/Qwen2-0.5B",temperature= 0.1, model_kwargs={"max_length":64})
```

```python
#predict something
text = "What is capital of India"
llm(prompt=text)
```

```python
from typing import Annotated
from typing_extensions import TypedDict
from langgraph.graph import StateGraph, START,END
from langgraph.graph.message import add_messages
```

```python
class State(TypedDict):
    message:Annotated[list,add_messages]
```

```python
#creating the Graph Builder
graph_builder = StateGraph(State)
graph_builder
```

```python
#Defining the Chatbot
def chatbot(state:State):
  return{"message":llm.revoke(state["message"])}
```

```python
#Adding the node to Graph Builder
graph_builder.add_node("chatbot",chatbot)
```

```python
#adding the start and end nodes
graph_builder.add_edge(START,"chatbot")
graph_builder.add_edge("chatbot",END)
```

```python
#Compiling the graph
graph = graph_builder.compile()
```

```python
user_ip = input("Enter your question: ")
for event in graph.stream({'message':("user",user_ip)}):
  print(event.values())
  for value in event.values():
    print(value['messages'])
    print(value['messages'].content)
```
